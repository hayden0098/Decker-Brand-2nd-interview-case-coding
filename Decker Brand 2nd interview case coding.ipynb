```python
def json_to_comma_separated(json_str):
    try:
        dict_list = json.loads(json_str)
        return ', '.join([item.get('name', item.get('id', '')) for item in dict_list])
    except (json.JSONDecodeError, TypeError):
        return ''

# 处理 DataFrame 中的指定列
columns_to_process = ['genres', 'keywords', 'production_companies', 'production_countries', 'cast', 'crew']

for column in columns_to_process:
    merge_df[column] = merge_df[column].apply(json_to_comma_separated)
tmdb_5000_movies_df
```
